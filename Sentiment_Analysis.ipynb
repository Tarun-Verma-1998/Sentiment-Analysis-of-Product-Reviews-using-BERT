#### importing BertTokenizer for tokenising to textual reviews

```python
from transformers import BertTokenizer
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
dataFrame['bert_tokenized_text'] = dataFrame['text'].apply(lambda x: tokenizer.tokenize(x))
```

```python
from keras.preprocessing.sequence import pad_sequences
```

```python
# Converting tokenized text to token IDs
dataFrame['bert_tokenized_text_ids'] = dataFrame['bert_tokenized_text'].apply(lambda x: tokenizer.convert_tokens_to_ids(x))
```

```python
# Padding the tokenized sequences to a length of 512 which is suitable for BERT input
max_length = 512
padded_sequences = pad_sequences(dataFrame['bert_tokenized_text_ids'], maxlen=max_length, padding='post', truncating='post')
dataFrame['padded_sequences'] = list(padded_sequences)
```

```python
dataFrame.head()
```

```python
# Adding special tokens [CLS] and [SEP]
padded_sequences_with_special_tokens = []
for seq in dataFrame['padded_sequences']:
    padded_sequences_with_special_tokens.append([tokenizer.cls_token_id] + seq + [tokenizer.sep_token_id])
```

```python
# Creating segment IDs (all zeros for single-sequence classification)
segment_ids = [[0] * max_length for _ in range(len(dataFrame))]
```

```python
# Creating attention masks (1 for real tokens, 0 for padding tokens)
attention_masks = []
for seq in padded_sequences_with_special_tokens:
    mask = [1] * len(seq) + [0] * (max_length - len(seq))
    attention_masks.append(mask)
```

```python
# Updating the DataFrame with the formatted sequences
dataFrame['input_ids'] = padded_sequences_with_special_tokens
dataFrame['segment_ids'] = segment_ids
dataFrame['attention_masks'] = attention_masks
```

```python
dataFrame.head()
```

```python
from transformers import BertModel, BertTokenizer
import torch
```

```python
from transformers import TFBertModel, BertTokenizer
model = TFBertModel.from_pretrained('bert-base-uncased')
```

```python
# Converting input data to TensorFlow tensors
import tensorflow as tf
input_ids_tf = tf.convert_to_tensor(dataFrame['input_ids'].tolist())
attention_masks_tf = tf.convert_to_tensor(dataFrame['attention_masks'].tolist())
segment_ids_tf = tf.convert_to_tensor(dataFrame['segment_ids'].tolist())
```

```python
import numpy as np

batch_size = 32  # Choose an appropriate batch size

# Obtain contextual embeddings in batches
num_batches = 4 #len(input_ids_tf) // batch_size
all_sentiment_scores = []
np.set_printoptions(precision=4, suppress=True)

for i in range(num_batches):
    # Calculate the start and end index for the current batch
    start_idx = i * batch_size
    end_idx = min((i + 1) * batch_size, len(dataFrame))

    # Get a batch of input data
    input_ids_batch = input_ids_tf[start_idx:end_idx]
    attention_masks_batch = attention_masks_tf[start_idx:end_idx]
    segment_ids_batch = segment_ids_tf[start_idx:end_idx]

    # Obtain contextual embeddings
    with tf.device('/cpu:0'):  # Use CPU for tokenization
        outputs = model(input_ids_batch, attention_mask=attention_masks_batch, token_type_ids=segment_ids_batch)

    # Get the contextual embeddings for the first token ([CLS] token)
    embedding = outputs.last_hidden_state[:, 0, :]

    # Add a classification head (linear layer without activation) to predict sentiment scores
#     classification_head = tf.keras.layers.Dense(1)
    classification_head = tf.keras.layers.Dense(1, activation='tanh')
    sentiment_scores = classification_head(embedding)

    # Scale the sentiment scores to the range [-1, 1]
    sentiment_scores_scaled = 2 * tf.keras.activations.tanh(sentiment_scores)

    # Append the sentiment scores to the list
    all_sentiment_scores.append(sentiment_scores_scaled.numpy())

# Convert the list of sentiment scores to a numpy array
all_sentiment_scores = np.concatenate(all_sentiment_scores, axis=0)

# Print the scaled sentiment scores for the entire dataset
print((all_sentiment_scores))
```
